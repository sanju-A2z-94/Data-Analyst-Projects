# Line Plot :-

```python
plt.plot(x,y)
plt.title("My line plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

```python
plt.plot(x,y, color='g', linestyle=':', linewidth=5)
plt.title("My line plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

***Adding two lines in the same plot***

```python
y2 = 4*x 
```

```python
plt.plot(x,y, color='g', linestyle=':', linewidth=5)
plt.plot(x,y2, color='r', linestyle='-.', linewidth=5)
plt.title("Line Plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.grid(True)
plt.show()
```

```python
#Adding Sub-plots---
```

```python
plt.subplot(2,1,1)
plt.plot(x,y, color='g', linestyle=':', linewidth=5)

plt.subplot(2,1,2)
plt.plot(x,y2, color='r', linestyle='-.', linewidth=5)

plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y, color='g', linestyle=':', linewidth=5)

plt.subplot(1,2,2)
plt.plot(x,y2, color='r', linestyle='-.', linewidth=5)

plt.show()
```

# Bar Plot :-

```python
#Used for categorical value-----

Employee = {"David":87000, "Andy":145000, "Sanjoy":280000}
```

```python
Names = list(Employee.keys())
```

```python
Salary = list(Employee.values())
```

```python
plt.bar(Names, Salary)
plt.show()
```

```python
plt.bar(Names, Salary, color='greenyellow')
plt.title("Bar plot of employee-salary")
plt.xlabel("Names of Employee")
plt.ylabel("Salary of Employee")
plt.grid(True)
plt.show()
```

```python
#Horizontal Bar Plot-----
```

```python
plt.barh(Names, Salary, color='mediumturquoise')
plt.title("Bar plot of employee-salary")
plt.xlabel("Salary of Employee")
plt.ylabel("Name of Employee")
plt.grid()
plt.show()
```

# Scatter Plot :-

```python
#Basic Scatter Plot-----
```

```python
r=[10,20,30,40,50,60,70,80,90]
s=[4,7,2,9,6,10,12,8,11]
```

```python
plt.scatter(r,s, color='indigo')
plt.title("Scatter Plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

```python
#Changing Mark Aesthetics-----
```

```python
plt.scatter(r,s, marker='h', color='darkmagenta', s=150)
plt.title("Scatter Plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

***Adding two Scatter Plot***

```python
t=[6,9,4,11,8,12,14,10,13]
```

```python
plt.scatter(r,s, marker='h', color='darkmagenta', s=150)
plt.scatter(r,t, marker='d', color='darkcyan', s=120)
plt.title("Two Scatter Plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.grid(True)
plt.show()
```

```python
#Adding Sub-Plots-----
```

```python
plt.subplot(1,2,1)
plt.scatter(r,s, marker='h', color='darkmagenta', s=150)
plt.grid(True)

plt.title("Scatter Plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")

plt.subplot(1,2,2)
plt.scatter(r,t, marker='d', color='darkcyan', s=120)
plt.grid(True)

plt.title("Scatter Plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")

plt.show()
```

```python
r_1 = np.linspace(0,5,10)
s_1 = r_1**2
plt.plot(r_1, s_1)
plt.title('Days Squared chart')
plt.xlabel('Days')
plt.ylabel('Days Squared')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(r_1, s_1, 'cyan')

plt.subplot(1,2,2)
plt.plot(r_1, s_1, 'green')

plt.show()
```

# Using Figure Objects

```python
fig_1 = plt.figure(figsize=(5,4), dpi=100)
axes_1 = fig_1.add_axes([0.1,0.1,0.9,0.9])
axes_1.set_xlabel('Days')
axes_1.set_ylabel('Days Squared')
axes_1.set_title('Days Squared Chart')
axes_1.plot(r_1, s_1, label='r/r^2')
axes_1.plot(s_1, r_1, label='r^2/r')
axes_1.legend(loc=0)

axes_2 = fig_1.add_axes([0.45, 0.45, 0.4, 0.3])
axes_2.set_xlabel('Days')
axes_2.set_ylabel('Days Squared')
axes_2.set_title('Days Squared Chart')
axes_2.plot(r_1, s_1, 'green')
axes_2.text(0, 40, 'Working')

plt.show()
```

# SubPlots:-

```python
fig_2, axes_2 = plt.subplots(figsize=(10,5), nrows=1, ncols=3)
plt.tight_layout()
axes_2[1].set_title('Days Squared Chart')
axes_2[1].set_xlabel('Days')
axes_2[1].set_ylabel('Days Squared')
axes_2[1].plot(r_1, s_1)

plt.show()
```

```python
fig_3 = plt.figure(figsize=(7,5))
axes_3 = fig_3.add_axes([0,0,1,1])
axes_3.set_title('Days Squared Chart')
axes_3.set_xlabel('Days')
axes_3.set_ylabel('Days Squared')
axes_3.plot(r_1,s_1, color='teal', alpha=.75, lw=2, ls='-.', marker='d', markersize=23,markerfacecolor='hotpink', markeredgecolor='mediumpurple', markeredgewidth=5)


axes_3.grid(True, color='0.6', dashes=(5,2,1,2))
axes_3.set_facecolor('#FAEBD7')

plt.show()
```

```python
np_arr = Air_Conditioner_Sales.values
p_1 = np_arr[:,0]
q_1 = np_arr[:,1]
fig_4 = plt.figure(figsize=(5,4), dpi=150)
axes_4 = fig_4.add_axes([0,0,1,1])

axes_4.set_title('Air Conditioner Actual Price vs. Ratings')
axes_4.set_xlabel('Ratings')
axes_4.set_ylabel('Actual Price of AC')
axes_4.plot(p_1,q_1, color='olive', alpha=.75, lw=3, ls='-.', marker='o', fillstyle='top',  markersize=8, markerfacecolor='teal', markeredgecolor='mediumaquamarine', markeredgewidth=2, label='ratings vs. actual_price')
axes_4.legend(loc=4)

axes_4.grid(True, color='0.6', dashes=(5,2,1,2))
axes_4.set_facecolor('greenyellow')
plt.show()
```
