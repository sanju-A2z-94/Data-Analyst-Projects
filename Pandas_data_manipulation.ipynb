# To find  a set of columns & rows--->

```python
newdf.loc[[7,9], ['B', 'D']]
```

```python
newdf.loc[[3,4],['D','E']]
```

```python
#All the rows but selected columns-----

newdf.loc[:,['B','D']]
```

```python
#All the columns but selected rows-----

newdf.loc[[3,5],:]
```

```python
newdf.loc[(newdf['A']<0.3)]
```

```python
newdf.loc[(newdf['C']>0.4)]
```

```python
#Returning common rows having required values-----

newdf.loc[(newdf['A']<0.3) & (newdf['C']>0.4)]
```

```python
newdf.head(3)
```
